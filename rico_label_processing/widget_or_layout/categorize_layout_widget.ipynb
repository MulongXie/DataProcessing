```python
import pandas as pd
```

```python
class_all = pd.read_csv('rico_class_org.csv', index_col=0)
```

```python
layout_str = 'Layout|PhoneWindow|Container|Card|view.View|BottomBar|'\
            + 'webview|WebView|Toolbar|ListView|ScrollView|RecyclerView|'\
            + 'GridView|TableRow|AdView|ads.internal.bp|NavigationMenu|ads.internal.bi|'\
            + 'ListMenuItemView|RadioGroup|ViewFlipper'
filter_layout = class_all['class_name'].str.contains(layout_str)
```

```python
widget_str = 'Text|Image|Button|ActionMenu|ProgressBar|UrlInputView|v7.widget.ab' \
            + '|CheckBox|ActionBarView|Spinner|SwitchCompat' \
            + '|widget\.\w{1,4}$'
filter_widget = class_all['class_name'].str.contains(widget_str)
```

```python
class_layout = class_all[filter_layout]
class_layout['category'] = 'layout'
class_widget = class_all[filter_widget]
class_widget['category'] = 'widget'
class_select = pd.concat([class_layout, class_widget]).sort_index()
class_left = class_all[~ (filter_layout | filter_widget)]
```

```python
class_select['number'].sum() / class_all['number'].sum()
```

```python
class_select.to_csv('rico_class_widget_layout.csv')
```
