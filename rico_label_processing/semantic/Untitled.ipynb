```python
import pandas as pd
import json
from tqdm import tqdm
```

```python
label2class = json.load(open('label_to_class.json'))
class_all = pd.read_csv('rico_class_org.csv', index_col=0)
class_all['label_name'] = None
```

```python
for i in tqdm(range(len(class_all))):
    class_name = class_all.loc[i,'class_name']
    for label in label2class:
        if class_name in label2class[label]:
            class_all.loc[i,'label_name'] = label
```
